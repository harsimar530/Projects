```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
df=yf.download("FDX",start='2020-01-01',end='2022-07-23')
df=df.drop(columns=['Adj Close'])
```

```python
df['SMA']=df['Close'].rolling(window=5).mean()
df['LMA']=df['Close'].rolling(window=20).mean()
df['IBS']=(df['Close']-df['Low'])/(df['High']-df['Low'])

days=[0]*len(df['Close'])
for i in range(0,len(df['Close'])):
    for j in range(i,len(df['Close'])):
        if (df['Close'].iloc[j]-df['Close'].iloc[i])/df['Close'].iloc[i] >= .07:
            cnt=j-i
            days[i]=cnt
            break
        else:
            continue                
        
df['time2']=days
df['gain']=np.where(df['Close'] > df['Close'].shift(1),df['Close']/df['Close'].shift(1),0)
df['loss']=np.where(df['Close'] < df['Close'].shift(1),df['Close']/df['Close'].shift(1),0)
df['avg_gain']=df['gain'].rolling(window=20).mean()
df['avg_loss']=df['loss'].rolling(window=20).mean()
df['RS']=df['avg_gain']/df['avg_loss']
df['RSI']=100-((100)/(1+df['RS']))
```

```python
x=[]
x=list(range(1,25))

for i in x:
    AG=df['gain'].rolling(window=i).mean()
    AL=df['loss'].rolling(window=i).mean()
    RS= AG/AL
    RSI=100-((100)/(1+RS))
    df.loc[:, str(i)]= RSI
    
```
